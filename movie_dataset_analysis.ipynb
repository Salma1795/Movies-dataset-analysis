```python
#na_values parameter is used to tell pandas they consider "not available" as NaN value and print NaN at the place of "not available".
df2 = pd.read_csv('/content/drive/MyDrive/movie dataset/movie_statistic_dataset.csv',na_values=["-"])
df2
```

```python
#count the null values
null_values = df2.isnull().sum()
null_values
```

```python
# to see null values percentage from the entire data.
#about 2% of the dataset is null
import numpy as np
data_count = np.product(df2.shape)
(null_values.sum()/data_count)*100
```

```python
# drop the null values, which will drop all row with a null value
df2 = df2.dropna(axis = 0)
df2
```

# **runtime_minutes vs movie_averageRating**

```python
# runtime_minutes vs movie_averageRating

from matplotlib import pyplot as plt
df2.plot(kind='scatter', x='runtime_minutes', y='movie_averageRating', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Correlation
# result: The correlation rate is closer to 0, making the linear association weaker.
#As runtime increases, the average rating tends to increase, but this relationship is not perfect or very strong.
correlation = df2['runtime_minutes'].corr(df2['movie_averageRating'])
print("Correlation result between runtime and average rating:", correlation)
```

# **movie_averageRating vs movie_numberOfVotes**

```python
# movie_averageRating vs movie_numerOfVotes

df2.plot(kind='scatter', x='movie_averageRating', y='movie_numerOfVotes', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Correlation
# result: The correlation rate is closer to 0, making the linear association weaker.
# this correlation is indicating a moderate positive correlation.
#This suggests that as the number of votes increases, the average rating also tends to increase, but the relationship is not very strong.
correlation = df2['movie_averageRating'].corr(df2['movie_numerOfVotes'])
print("Correlation result between number of votes and average rating:", correlation)
```

```python
# Domestic gross $ vs Worldwide gross $
df2.plot(kind='scatter', x='Domestic gross $', y='Worldwide gross $', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#The Pearson correlation coefficient of 0.9377 between Domestic Gross and Worldwide Gross indicates a strong positive correlation.
# result: The correlation rate is closer to 1, making the linear association strong.
correlation = df2['Domestic gross $'].corr(df2['Worldwide gross $'])
print("Correlation result between Domestic gross and Worldwide gross:", correlation)
```
